```python
flow={}
count=0
for i, j in df.iterrows(): 
    count+=1
    t=(j.src,j.dst,j.sport,j.dport,j.proto,j.macsrc,j.macdst) 
    if t in flow:
        flow[t].append([j.time,j.len])
    else:
        flow[t]=[[j.time,j.len]]
    if count%10000==0:
        print(count,end="\r")
```

```python
import pickle

with open('flow.pickle', 'wb') as handle:
    pickle.dump(flow, handle, protocol=pickle.HIGHEST_PROTOCOL)

with open('flow.pickle', 'rb') as handle:
    b = pickle.load(handle)

print(flow == b)
```
